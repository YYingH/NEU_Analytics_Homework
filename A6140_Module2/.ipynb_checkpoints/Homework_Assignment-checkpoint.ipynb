#### 1. Retrieve the data for each series

```python
start_date = '2000-01-01'
end_date = '2019-11-17'
sp500_data = data.get_data_yahoo('^GSPC', start_date, end_date, interval='m')
sp500_data['DATE']=sp500_data.index
sp500_data['DATE'] = pd.to_datetime(sp500_data['DATE'])
sp500_data['SP500'] = sp500_data['Adj Close'].astype('uint32')
```

```python
truckd11_data = pd.read_csv("Data/TRUCKD11.csv", na_values=".").dropna()
truckd11_data['DATE'] = pd.to_datetime(truckd11_data['DATE'])
recession_data = pd.read_excel("Data/Recession.xlsx", na_values = ".").dropna()
recession_data['Peak'] = pd.to_datetime(recession_data['Peak'])
recession_data['Trough'] = pd.to_datetime(recession_data['Trough'])
```

```python
sp500_data.plot(x="DATE", y = "SP500", label = "SP500", figsize = (20,5))
plt.xlabel("time")
plt.ylabel("SPY stuals")
plt.legend(loc="best")
plt.title("SPY500 Dataset")
plt.show()

truckd11_data.plot(x="DATE", y="TRUCKD11", label="TRUCKD11", figsize = (20,5))
plt.xlabel("time")
plt.ylabel("TRUCKD11")
plt.legend(loc="best")
plt.title("TRUCKD11 Dataset")
plt.show()
```

```python
fig = plt.figure(figsize = (40,15))

ax1 = fig.add_subplot(111)
ax1.plot(truckd11_data["DATE"], truckd11_data["TRUCKD11"],'-r', label='TRUCKD11')
# ax1.set_xlim([min(truckd11_data["DATE"]), max(truckd11_data["DATE"])])
plt.legend(loc='upper left')
ax2 = ax1.twinx()
ax2.plot(sp500_data["DATE"], sp500_data["SP500"],'-', label = 'SP500')
# ax2.set_xlim([min(sp500_data["DATE"]), max(sp500_data["DATE"])])

plt.legend(loc='upper right')

plt.show()
```

#### 2. Apply a log transform to the data (where appropriate). 

```python
sp500_data_log = sp500_data.copy()
sp500_data_log['SP500'] = np.log(sp500_data_log['SP500'])
sp500_data_log.plot(x="DATE", y = "SP500", label = "SP500", figsize = (20,5))
plt.xlabel("time")
plt.ylabel("SPY stuals")
plt.legend(loc="best")
plt.title("SPY500 Dataset")
plt.show()

truckd11_data_log = truckd11_data.copy()
truckd11_data_log['TRUCKD11'] = np.log(truckd11_data_log['TRUCKD11'])
truckd11_data_log.plot(x="DATE", y="TRUCKD11", label="TRUCKD11", figsize = (20,5))
plt.xlabel("time")
plt.ylabel("TRUCKD11")
plt.legend(loc="best")
plt.title("TRUCKD11 Dataset")
plt.show()
```

#### 3. Plot the data and identify any relationships

```python
fig = plt.figure(figsize = (40,15))

ax1 = fig.add_subplot(111)
ax1.plot(truckd11_data_log["DATE"], truckd11_data_log["TRUCKD11"],'-r', label='TRUCKD11')
# ax1.set_xlim([min(truckd11_data["DATE"]), max(truckd11_data["DATE"])])
plt.legend(loc='upper left', prop={'size': 20})
ax2 = ax1.twinx()
ax2.plot(sp500_data_log["DATE"], sp500_data_log["SP500"],'-', label = 'SP500')
# ax2.set_xlim([min(sp500_data["DATE"]), max(sp500_data["DATE"])])

plt.legend(loc='upper right', prop={'size': 20})

plt.show()
```

#### 4. Shade areas that contain recessions

```python
recession_data_new = recession_data.loc[(recession_data['Peak']>= start_date)]
fig = plt.figure(figsize = (40,15))

ax1 = fig.add_subplot(111)
ax1.plot(truckd11_data_log["DATE"], truckd11_data_log["TRUCKD11"],'-r', label='TRUCKD11')

plt.legend(loc='upper left', prop={'size': 20})
ax2 = ax1.twinx()
ax2.plot(sp500_data_log["DATE"], sp500_data_log["SP500"],'-', label = 'SP500')

for _, rows in recession_data_new.iterrows():
    plt.axvspan(rows['Peak'],rows['Trough'], color='gray',alpha=0.3)
plt.legend(loc='upper right', prop={'size': 20})
plt.show()
```

### Part. 2  - An economist asserts that stock market direction is largely based on whether the S&P 500 Index is above/below the 200 Day Simple Moving Average (SMA).

- [x] Retrieve the data for the S&P 500
- [x] Create simple moving average columns for not only the 200 day SMA, but also for other popular SMAs (e.g., 20, 50, and 100)
- [ ] Plot the data and identify any relationships
- [x] Shade areas that contain recessions
- [ ] Explain how the information may or may not be used


```python
start_date = '1996-01-01'
end_date = '2019-11-17'
recession_data_new = recession_data.loc[(recession_data['Peak']>= start_date)]
sp500_data = data.get_data_yahoo('^GSPC', start_date, end_date, interval='d')
sp500_data['SP500'] = sp500_data['Adj Close'].astype('uint32')
sp500_data['DATE']=sp500_data.index

sp500_data_log = sp500_data.copy()
sp500_data_log['SP500'] = np.log(sp500_data['SP500'])
sp500_data_log.plot(x="DATE", y = "SP500", label = "SP500", figsize = (20,5), lw=0.5)
plt.xlabel("time")
plt.ylabel("SPY stuals")

for sp in [20,50,100,200]:
    rolling_mean = sp500_data_log.Open.rolling(window=sp).mean()
    lb = 'AMD '+ str(sp) +' Day SMA'
    plt.plot(sp500_data_log.DATE, rolling_mean, label = lb, lw=0.6)
for _, rows in recession_data_new.iterrows():
    plt.axvspan(rows['Peak'],rows['Trough'], color='gray',alpha=0.3)

    
plt.legend(loc="best")
plt.title("SPY500 Dataset")
plt.show()
```
