### Part 3. Hypothesis Testing (25%)

Analyze both data sets and determine if either marketing campaign was successful in raising the average miles driven per Uber driver.

- The Null Hypothesis is that the average miles driven per Uber driver in marketing campaign is as same as not in marketing campaign.

- The Alternate Hypothesis is that the average miles driven per Uber driver will be increased when in the period of marketing campaign.

- The p-value will be set to .05

```python
data0 = lax_df_trimmed_cleanned['LAX - Avg. Daily Miles Driven Per Hour'].loc[lax_df_trimmed_cleanned['Marketing_Campaign']==0]
data1 = lax_df_trimmed_cleanned['LAX - Avg. Daily Miles Driven Per Hour'].loc[lax_df_trimmed_cleanned['Marketing_Campaign']==1]
plt.figure(figsize=(18, 6))
ax = sns.distplot(data0, hist=True)
ax = sns.distplot(data1, hist=True)
plt.legend(['No Marketing Campaign','Marketing Campaign'])
```

```python
# compare samples
stat, p = f_oneway(data0, data1)
print('Statistic = %.3f, p=%.3f'%(stat, p))

# interpret
alpha = 0.05
if p > alpha:
    print('Same distributions (fail to reject H0)')
else:
    print('Different distributions (reject H0)')
```

```python
data2 = sfo_df_trimmed_cleanned['LAX - Avg. Daily Miles Driven Per Hour'].loc[sfo_df_trimmed_cleanned['Marketing_Campaign']==0]
data3 = sfo_df_trimmed_cleanned['LAX - Avg. Daily Miles Driven Per Hour'].loc[sfo_df_trimmed_cleanned['Marketing_Campaign']==1]
plt.figure(figsize=(18, 6))
ax = sns.distplot(data2, hist=True)
ax = sns.distplot(data3, hist=True)
plt.legend(['No Marketing Campaign','Marketing Campaign'])
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(23,8))
sns.set(style='whitegrid')
ax = sns.distplot(data0, hist=True, kde_kws={"label": "LAX:0"})
ax = sns.distplot(data1, hist=True, kde_kws={"label": "LAX:1"})
ax = sns.distplot(data2, hist=True, kde_kws={"label": "SFO:0"})
ax = sns.distplot(data3, hist=True, kde_kws={"label": "SFO:1"})
```

```python
# compare samples
stat, p = f_oneway(data2, data3)
print('Statistic = %.3f, p=%.3f'%(stat, p))

# interpret
alpha = 0.05
if p > alpha:
    print('Same distributions (fail to reject H0)')
else:
    print('Different distributions (reject H0)')
```

```python
# compare samples
stat, p = f_oneway(data0, data1, data2, data3)
print('Statistic = %.3f, p=%.3f'%(stat, p))

# interpret
alpha = 0.05
if p > alpha:
    print('Same distributions (fail to reject H0)')
else:
    print('Different distributions (reject H0)')
```

**Finding**

- The Null hypothesis has been reject because of the P-value is less than (or equal to) 0.05, and we will accept the alternative hypothesis that the average miles driven per Uber driver will be increased when in the period of marketing campaign

**What factors may affect the validity of this study?**
There are three factors that might affected the power of this hypothesis test:

- Sample size ($n$). Other things being equal, the greater the power of the test.

- Significance leval ($\alpha$). The lower the significance level, the lower the power of the test. If you reduce the significance level (e.g., from 0.05 to 0.01, the region of acceptance gets bigger. As a result, you are less likely to reject the null hypothesis. This means you are less likely to reject the null hypothesis when it is false, so you are more likely to make a Type II error. In short, the power of the test is reduced when you reduce the significance level; and vice versa.

- The "true" value of the parameter being tested. The greater the difference between the "true" value of a parameter and the value specified in the null hypothesis, the greater of the test. That is, the greater the effect size, the greater the power of the test.
